### Several methods to generate a DataFrame
1. Import from Excel/CSV.
2. Build up manually.
3. Generate randomly by python default syntax.
4. Import from online address.

***Besides,***
- If want to keep the origin DataFrame, `copy` a new DataFrame beforehand then start manipulating could be a good way.

```python
import pandas as pd
```

# Import from Excel/CSV

```python
df = pd.read_csv('data_from_trip_practice_only.csv')
```

```python
df.head()
```

# Build up manually
using Python `Dict`

```python
dic = {
    "col 1": [1, 2, 3], 
    "col 2": [10, 20, 30],
    "col 3": list('xyz'),
    "col 4": ['a', 'b', 'c'],
    "col 5": pd.Series(range(3))
}
df1 = pd.DataFrame(dic)
df1
```

# Generate randomly by python default syntax.

***Include all data types.***

```python
df2 = pd.util.testing.makeMixedDataFrame()
df2
```

***Include Null data.***

```python
df3 = pd.util.testing.makeMissingDataframe().head()
df3
```

***With `Date` type index***

```python
df4 = pd.util.testing.makeTimeDataFrame().head(10)
df4
```
