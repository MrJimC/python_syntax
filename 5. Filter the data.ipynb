## Basic methods
- `df['column_name']`
- `iloc` / `loc`

```python
df_titanic = pd.read_csv('http://bit.ly/kaggletrain')

# 複製一份副本 DataFrame
df_titanic_copy = df_titanic.copy()
df_titanic_copy
```

***use `df['column_name']`***

```python
df_titanic_copy['Name']
```

```python
df_titanic_copy[['Name','Sex']]
```

```python
get_column = ['Name','Sex']
df_titanic_copy[get_column]
```

***use `iloc`***

```python
df_titanic_copy.iloc[:2,:5]
```

```python
df_titanic_copy.iloc[[0, 1, 2], [0,1,2,3,4,5]]
```

```python
df_titanic_copy.iloc[-5:]
```

***use `loc`***

```python
df_titanic_copy.loc[0, 'Name']
```

```python
df_titanic_copy.loc[0:5, ['Name','Age']]
```

```python
df_titanic_copy.loc[df_titanic_copy.Sex == 'male']
```

```python
df_titanic_copy.loc[(df_titanic_copy.Sex == 'male') & (df_titanic_copy.Age >= 50)].head(6)
```

```python
df_titanic_copy.loc[(df_titanic_copy.Sex == 'male') | (df_titanic_copy.Age >= 50)].head(6) 

# "|" means or
```

## Select columns with designated `data type`.

```python
df2 = pd.util.testing.makeMixedDataFrame()
df2
```

***eg. include columns with `number` data type only.***

```python
df2_1 = df2.select_dtypes(include='number').head()
df2_1
```

***eg. include columns which data type aren't `object`.***

```python
df2_2 = df2.select_dtypes(exclude=['object'])
df2_2
```
