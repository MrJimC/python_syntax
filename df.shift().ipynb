```python
import pandas as pd
```

```python
ind = pd.date_range('01 / 01 / 2000', periods = 5, freq ='12H') 
   
df = pd.DataFrame({"A":[1, 2, 3, 4, 5],  
                   "B":[10, 20, 30, 40, 50], 
                   "C":[11, 22, 33, 44, 55], 
                   "D":[12, 24, 51, 36, 2]},  
                    index = ind) 

df
```

---

### 索引軸向正方向移動2個周期的函數

```python
# 正方向
df.shift(2, axis = 0)
```

```python
# 負方向
df.shift(-2, axis = 0)
```

--- 

### 采用shift()用於在時間序列數據中將"列"軸移動2個周期

```python
df.shift(2, axis = 1)
```

```python
df.shift(-2, axis = 1)
```

### 實際使用場景
- 可能在計算 YoY, MoM...等場景較方便

```python
dfa = df.copy()
dfa
```

```python
dfa = dfa.drop(['B','C','D'], axis=1)
dfa
```

```python
dfa['A_last_period'] = dfa['A'].shift(1)
dfa
```

```python
dfa['change'] = dfa['A']-dfa['A_last_period']
dfa
```
