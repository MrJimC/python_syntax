```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_from_trip_practice_only.csv')
df.head()
```

---

### Basic use 1

```python
# 每個國家下，每個星級的酒店個數
pd.pivot_table(df, index=['countrynameen'], values=['masterhotelid'], columns=['star'], 
               aggfunc='count', margins=True)
```

```python
# 多個aggfunc聚合方式 (可寫入自定義函數)
pd.pivot_table(df, index=['countrynameen'], values=['this_wk_uv'], 
               aggfunc=[np.sum, np.mean, np.std, lambda x:max(x)], margins=True)
```

```python
# 多個index, columns的情況
pd.pivot_table(df, index=['countrynameen','lose_rate_tag'], values=['masterhotelid'], columns=['star','availability_tag'], 
               aggfunc='count', margins=True)
```
