# `df.groupby`


```python
import pandas as pd
```

```python
df = pd.read_csv('data_from_trip_practice_only.csv')
df.head()
```

---

## Basic use 1

```python
group1 = df.groupby('countrynameen')
```

```python
# group對象的內部
list(group1)
```

```python
group1.first()  # 取每個group的第一行
```

```python
group1.get_group('South Korea')  # 取group內的指定group值
```

```python
group1_multi = df.groupby(['countrynameen','star'])
group1_multi.first()
```

---

## 對不同欄位做不同操作

```python
df.groupby('countrynameen').agg({'this_wk_uv':'median','this_wk_order':'mean'})
```

## 套用自定義函數

```python
# 取每個國家下本週訂單最多的酒店

def get_highest_order_this_wk(x):
    df = x.sort_values(by = 'this_wk_order',ascending=True)
    return df.iloc[-1,:]

highest_order = df.groupby('countrynameen', as_index=False).apply(get_highest_order_this_wk)
highest_order
```
