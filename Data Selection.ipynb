```python
import pandas as pd
```

---

## Normal ways
- couldn't designate the Rows & Columns at the same time.

```python
# use only one bracket will return a List
df['countrynameen']
```

```python
# use double brackets will return a DataFrame
df[['countrynameen']]
```

```python
# pick rows by index.
df[1:3]
```

```python
# pick by columns names.
df[['masterhotelid','masterhotelname']]
```

```python
df[0:5, ['masterhotelid','masterhotelname']]
```

```python
df[0:3, 1:5]
```

---

## loc
- using column name to pick data.

```python
df.loc[1:3, ['countrynameen','masterhotelid']]
```

```python
df.loc[1:3, 'countrynameen':'masterhotelid']
```

```python
# could be used without defining the column name.
df.loc[1:3]
```

```python
# use only one bracket will return a List
df.loc[1]
```

```python
# use double brackets will return a DataFrame
df.loc[[1]]
```

```python
df.loc[df['this_wk_order']>100,['masterhotelid','masterhotelname']]
```

```python
df.loc[df['this_wk_order']>100,'masterhotelid':'financial_revenue_past_4w']
```

---

## iloc
- picking data by Index.

```python
# use only one bracket will return a List
df.iloc[0]
```

```python
# use double brackets will return a DataFrame
df.iloc[[0]]
```

```python
df.iloc[1:6, [1,2]]
```

```python
df.iloc[1:6, 1:6]
```

```python
# could be used without defining the column name.
df.iloc[1:5]
```

---

## at
- `at` to point specific data point.  `df.at[index, column_name]`
- 可理解為取單一值的 `loc`

```python
df.at[3, 'masterhotelid']
```

```python
# could only have one index num.
df.at[1:5, 'masterhotelid']
```

```python
# could only have one column name.
df.at[1, ['masterhotelid','masterhotelname']]
```

---

## iat
- `iat` to point specific data point. `df.iat[index, column_index]`
- 可理解為取單一值的 `iloc`

```python
df.iat[2, 3]
```

```python
# could only have one index num.
df.iat[1:2, 3]
```

```python
# could only have one column index.
df.iat[2, 3:6]
```

---

## ix
- `ix` to point specific data point. `df.ix[index, column_index]` / `df.ix[index, column_names]`

```python
df.ix[1:3, 0:5]
```

```python
df.ix[1:3, ['masterhotelid','masterhotelname']]
```

```python
# could be used by only designating the index.
df.ix[1:3]
```

## use condition

```python
df[df.this_wk_order <= 100]
```

```python
# if want to use more than 1 conditionm you should use 
    # logical_and / logical_or / logical_not / logical_xor
import numpy as np
condition = np.logical_and(df.this_wk_order <= 100 , df.this_wk_order >= 90)

df[condition]
```
